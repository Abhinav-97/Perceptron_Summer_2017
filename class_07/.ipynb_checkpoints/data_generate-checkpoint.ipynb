```python
def f(x):
    return 5*x**3 + 4*x - 3.0 + 50*(2*np.random.random()-1)
```

```python
k1 = np.asarray([f(x) for x in range(-1000, 1000)])
k2 = np.asarray([f(x) for x in range(-1000, 1000)])
```

```python
plt.scatter(range(-1000, 1000), k1)
plt.scatter(range(-1000, 1000), k2)
```
